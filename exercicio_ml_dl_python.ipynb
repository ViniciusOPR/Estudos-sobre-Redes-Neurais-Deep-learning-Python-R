```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
soja = pd.read_csv("soybean.csv")
soja
```

```python
previsores = soja.iloc[:,0:35].values
classe = soja.iloc[:,35].values
```

```python
for n in range(0, len(soja.columns)-1):
    if soja.dtypes[n] == object:
        labelencoder1 = LabelEncoder()
        previsores[:,n] = labelencoder1.fit_transform(previsores[:,n])
    
labelencoder_y = LabelEncoder()
classe = labelencoder_y.fit_transform(classe)
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(previsores, classe, test_size=0.3, random_state=0)
len(x_treinamento)
```

```python
x_treinamento = x_treinamento.astype('float32')
x_teste = x_teste.astype('float32')
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_treinamento, y_treinamento)
```

```python
previsoes = knn.predict(x_teste)
previsoes
```

```python
confusao = confusion_matrix(y_teste, previsoes)
confusao
```

```python
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_erro = 1 - taxa_acerto
taxa_acerto
```
